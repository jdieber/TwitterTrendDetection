```python
def get_new_tweets():
    
    master_list= []
    try:
        with open(file_name, 'r') as myfile:
            data = json.loads(myfile.read())
    except JSONDecodeError:
        pass

    for i in range (len(data)):
        master_list.append(data[i])
        
    names = [ 
                            '@cnnbrk', 
                            '@MSNBC',
                            '@GretaThunberg',
                            '@realDonaldTrump',
                            '@BarackObama', 
                            '@nytimes',
                            '@washingtonpost',
                            '@NPR',
                            '@AJEnglish',
                            '@FoxNews',
                            '@BBCBreaking',
                            '@Reuters'
                                        ]
    for i in names:
        timeline = api.user_timeline(screen_name=i, count=10, tweet_mode = 'extended',include_rts = True)
        for tweets in range(len(timeline)):
            id = timeline[tweets]._json['id']
            name = timeline[tweets]._json['user']['screen_name']
            text = timeline[tweets]._json['full_text']
            created_at =str(datetime.strptime(timeline[tweets]._json['created_at'], '%a %b %d %H:%M:%S +0000 %Y'))
            retweets = timeline[tweets]._json['retweet_count']
            likes = timeline[tweets]._json['favorite_count']
            followers = timeline[tweets]._json['user']['followers_count']
            time_interval = int(time.time())
            hashtags_list = []
            for i in range(len(timeline[tweets]._json['entities']['hashtags'])):
                hashtags_list.append(timeline[tweets]._json['entities']['hashtags'][i]['text'])

            userMentions_list = []    
            for i in range(len(timeline[tweets]._json['entities']['user_mentions'])):
                userMentions_list.append(timeline[tweets]._json['entities']['user_mentions'][i]['screen_name'])

            hashtags = hashtags_list
            user_mentions = userMentions_list
            if any(d['id'] == id for d in master_list):
                for i in range(len(master_list)):
                    if master_list[i]['counter'] >= 60:
                        pass
                    
                    elif id == master_list[i]['id'] and master_list[i]['counter'] < 60:
                        master_list[i]['counter'] = master_list[i]['counter'] + 1
                        master_list[i]['update ' + str(master_list[i]['counter'])] = int(time.time())
                        master_list[i]['likes ' + str(master_list[i]['counter'])] = likes
                        master_list[i]['retweets ' + str(master_list[i]['counter'])] = retweets
            else:
                tweet = {}
                counter = 0
                tweet.update(id = id,
                             counter=counter,
                             name = name,
                             text = text,
                             created_at = created_at,
                             time_interval = time_interval,
                             retweets = retweets,
                             likes = likes,
                             followers = followers,
                             hashtags = hashtags,
                             user_mentions = user_mentions)
                master_list.append(tweet) 
                
    with open(file_name, 'w') as fout:
        json.dump(master_list, fout)

    master_list = []        
```
